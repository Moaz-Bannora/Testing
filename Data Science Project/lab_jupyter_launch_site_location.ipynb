```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

Are all launch sites in proximity to the Equator?
Not exactly, but many are strategically closer to the Equator than other global options. Here's why that matters:
- Physics advantage: The Earth rotates fastest at the Equator, giving rockets a natural velocity boost. This reduces fuel requirements for reaching orbit.
- SpaceX examples:
- Cape Canaveral, Florida: ~28°N latitude—not equatorial, but still relatively close.
- Kennedy Space Center: Same region, similar benefits.
- Boca Chica (Starbase), Texas: ~26°N latitude—closer to the Equator than most U.S. sites.
So while none of SpaceX’s sites are on the Equator, they’re positioned to take advantage of its proximity for orbital efficiency.

Are all launch sites in very close proximity to the coast?
Yes—coastal access is a consistent feature across SpaceX’s launch sites. Here's why:
- Safety: Launching over open water minimizes risk to populated areas in case of failure.
- Trajectory flexibility: Coastal sites allow rockets to launch eastward over the ocean, aligning with Earth’s rotation.
- Examples:
- Cape Canaveral: Atlantic coast of Florida.
- Vandenberg Space Force Base: Pacific coast of California.
- Boca Chica (Starbase): Gulf Coast of Texas.
This coastal positioning isn’t just convenient—it’s essential for safe and efficient launch operations.

Summary of Findings
- SpaceX launch sites are not equatorial, but they’re close enough to benefit from Earth’s rotation.
- All sites are coastal, which supports safety, trajectory planning, and logistical access.
- These choices reflect a balance between physics, geography, and operational needs—a great example of how data and location strategy intersect.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from math import sin, cos, sqrt, atan2, radians
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Create marker color column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Site map setup
nasa_coordinate = [28.5623, -80.5772]  # Example center point
site_map = folium.Map(location=nasa_coordinate, zoom_start=14)
marker_cluster = MarkerCluster()

# Add cluster to map
site_map.add_child(marker_cluster)

# Add launch outcome markers
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color=record['marker_color']),
        popup=f"{record['Launch Site']} | {'Success' if record['class'] == 1 else 'Failure'}"
    )
    marker_cluster.add_child(marker)

# Coastline proximity example
launch_site_lat = 28.56321
launch_site_lon = -80.57672
coastline_lat, coastline_lon = 28.5624, -80.56766
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Coastline distance marker
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(50, 20),
        icon_anchor=(25, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
    )
)
marker_cluster.add_child(distance_marker)

# Show the map
site_map
```

```python
import folium
from math import sin, cos, sqrt, atan2, radians
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Define the haversine formula for distance calculation
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Add marker color column to launch records
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Define launch site (example: CCAFS SLC-40)
launch_site_lat, launch_site_lon = 28.5623, -80.5772
launch_site = [launch_site_lat, launch_site_lon]

# Define nearby proximity targets
proximity_targets = {
    'Coastline 🌊': [28.5624, -80.56766],
    'Railroad 🛤️': [28.56335, -80.58691],
    'Highway 🛣️': [28.56401, -80.57086],
    'Cape Canaveral 🌆': [28.38846, -80.60574]
}

# Colors for each line
color_map = {
    'Coastline 🌊': 'purple',
    'Railroad 🛤️': 'darkred',
    'Highway 🛣️': 'orange',
    'Cape Canaveral 🌆': 'teal'
}

# Create base map centered on launch site
site_map = folium.Map(location=launch_site, zoom_start=14)

# Add MousePosition for coordinate tracking
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter
)
site_map.add_child(mouse_position)

# Initialize MarkerCluster
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# Add launch records to the cluster
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color=record['marker_color']),
        popup=f"{record['Launch Site']} | {'Success' if record['class'] == 1 else 'Failure'}"
    )
    marker_cluster.add_child(marker)

# Add proximity lines and distance labels
for label, point in proximity_targets.items():
    # Calculate distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, point[0], point[1])
    distance_str = f"{label} ({distance:.2f} km)"

    # Draw line from launch site to proximity point
    line = folium.PolyLine(
        locations=[launch_site, point],
        color=color_map[label],
        weight=2,
        opacity=0.8,
        tooltip=distance_str
    )
    site_map.add_child(line)

    # Optional: Add marker showing distance at destination point
    distance_marker = folium.Marker(
        location=point,
        icon=DivIcon(
            icon_size=(50, 20),
            icon_anchor=(25, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance)
        )
    )
    marker_cluster.add_child(distance_marker)

# Display the final interactive map
site_map
```
